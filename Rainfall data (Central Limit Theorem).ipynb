```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
rainfall = pd.read_csv("data_monthly_rainfall.csv")
rainfall.head(2)
plt.figure(figsize=(4,4))
sns.kdeplot(x="Rainfall", data=rainfall)
plt.show()
plt.figure(figsize=(4,4))
sns.histplot(rainfall["Rainfall"])
plt.show()
np.mean(rainfall["Rainfall"]),
len(rainfall)

sample_mean = []
for i in range(500):
    sample_data = []
    for i in range(1000):
        sample_data.append(np.random.choice(rainfall["Rainfall"]))
    sample_mean.append(np.random.choice(sample_data))
len(sample_mean)

plt.figure(figsize=(4,4))
sns.kdeplot(sample_mean)
plt.show()
```
