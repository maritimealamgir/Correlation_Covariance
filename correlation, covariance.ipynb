```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
SMI1 = pd.read_csv("F:/Study/Project/data.csv")
SMI2 = pd.read_csv("F:/Study/Project/test.csv")
SMI1.head(2)
SMI2.head(3)
sns.histplot(SMI1["Chl-a"])
SMI1.isnull().sum()
SMI1.info() #This command is for data type

SMI1_Corr = SMI1.select_dtypes(["float64"]).corr() 
#This command is for finding correlation of the selected data type. If we have 
#more data type, then we have to given the new type in the command box. 
#Such as SMI1.select_dtypes(["float64","int64"]).corr()

SMI1_Cov = SMI1.select_dtypes(["float64"]).cov()

#To get a heatmap of Correlation or covariance matrics, before running the following code, 
#we have to save the selected dara type as a new data.

plt.figure(figsize=(12,10))
sns.heatmap(SMI1_Corr, annot=True)
plt.show()

plt.figure(figsize=(12,10))
sns.heatmap(SMI1_Cov)
plt.show()
```
